```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

ChromeDriverManager().install()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
import time

driver.get('https://store.kyobobook.co.kr/bestseller/total/annual/tech')

time.sleep(3)
```

```python
items = []
#페이지링크
while True:
    try:
        for j in range(1,11):
            print(j)
            book = driver.find_element('xpath', f'/html/body/div/main/section/div/div/section/ol[1]/li[{j}]/div/div[2]/div[2]/a').text
            time.sleep(3)
            items.append(book)
        for j in range(1,11):
            print(j)
            book = driver.find_element('xpath',f'/html/body/div/main/section/div/div/section/ol[2]/li[{j}]/div/div[2]/div[2]/a').text
            time.sleep(3)
            items.append(book)

        #다음페이지 버튼
        driver.find_element('xpath',f'//*[@id="tabRoot"]/div[2]/div[1]/button[2]').click()
        time.sleep(5)
    except:
        break
items
```

```python
import pandas as pd

df = pd.DataFrame(items)
df.columns = ['제목']
df.to_csv('c:/data/book/kyobo-it-best.csv', index=False, encoding='ms949')
```

```python
#크롬 브라우저 종료
driver.quit()
```
