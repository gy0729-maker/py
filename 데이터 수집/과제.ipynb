```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://demowebshop.tricentis.com/books"

headers = {
    "User-Agent": "Mozilla/5.0"
}

res = requests.get(url, headers=headers)
res.raise_for_status()

soup = BeautifulSoup(res.text, "html.parser")
```

```python
books = []

items = soup.select("h2.product-title a")

for item in items:
    title = item.get_text(strip=True)
    books.append([title])

print(books)
```

```python
with open('c:/data/text/books.csv', 'w', newline='') as f:
    w = csv.writer(f)
    w.writerow(['title'])
    w.writerows(books)
```
