```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = 'https://finance.naver.com'
res = urlopen(url)
bs = BeautifulSoup(res, 'html.parser')
result1 = []
moneys = bs.select('#content > div.article2 > div.section1 > div.group1 > table > tbody > tr.down.bold > td:nth-child(2)')
print(moneys)
for money in moneys:
    result1.append(money.text.strip())
```

```python
result2 = []
exchange = bs.select('#content > div.article2 > div.section1 > div.group1 > table > tbody > tr > td:nth-child(2)')
print(exchange)
for ex in exchange:
    result2.append(ex.text.replace(',',' '))
result = list(zip(result1, result2))
print(result)
```
