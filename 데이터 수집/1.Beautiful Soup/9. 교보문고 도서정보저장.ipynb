```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib import parse

book_name = 'python'
book_name = parse.quote(book_name)
html = urlopen('https://search.kyobobook.co.kr/search?keyword=' + book_name + '&gbCode=TOT&target=kyobo')
bs = BeautifulSoup(html, 'html.parser')
```

```python
############################################
new_titles = []
new_authors = []
new_price = []
for i in range(1, 6):
    html = urlopen(
        'https://search.kyobobook.co.kr/search?keyword=' + book_name + '&gbCode=TOT&target=kyobo&page=' + str(i))
    print('https://search.kyobobook.co.kr/search?keyword=' + book_name + '&gbCode=TOT&target=kyobo&page=' + str(i))
    bs = BeautifulSoup(html, 'html.parser')

    titles = bs.select(
        '#shopData_list > ul > li > div.prod_area.horizontal > div.prod_info_box > div.auto_overflow_wrap.prod_name_group > div.auto_overflow_contents > div > a > span')

    for title in titles:
        if title.text == '[국내도서]' or title.text == '예약판매':
            continue
        new_titles.append(title.text)

    authors = bs.select(
        '#shopData_list > ul > li > div.prod_area.horizontal > div.prod_info_box > div.prod_author_info > div.auto_overflow_wrap.prod_author_group > div.auto_overflow_contents > div > a')

    for a in authors:
        new_authors.append(a.text)

    prices = bs.select(
        '#shopData_list > ul > li > div.prod_area.horizontal > div.prod_info_box > div.prod_price > span.price > span.val')

    for p in prices:
        new_price.append(p.text)
```

```python
############################################
books = list(zip(new_titles, new_authors, new_price))
print(len(new_titles), '권 검색')
print('=' * 50)
for book in books:
    print(book[0], '-', book[1], '-', book[2])
```

```python
############################################
import pandas as pd

df = pd.DataFrame(books, columns=['도서명', '저자', '정가'])
df.head()
```

```python
############################################
df.to_csv('c:/data/book/data.csv', encoding='ms949', index=False)
```
