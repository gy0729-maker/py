```python
from bs4 import BeautifulSoup
import urllib.request as req
import time

url = 'https://news.naver.com/section/101'
res = req.urlopen(url)
bs = BeautifulSoup(res, 'html.parser')
#_SECTION_HEADLINE_LIST_oqbme > li:nth-child(1) > div > div > div.sa_text > a > strong
a_list = bs.select('li > div > div > div.sa_text > a')
for a in a_list:
    print(a.text.strip())
    print(a['href'])
    time.sleep(0.1)
```
