```python
import pandas as pd

df = pd.read_csv('c:/data/insta/insta_test.csv')
df.head()
```

```python
#더미변수
gender_dummy = pd.get_dummies(df['gender'], prefix='gender')
income_level_dummy = pd.get_dummies(df['income_level'], prefix='income')
employment_status_dummy = pd.get_dummies(df['employment_status'],prefix='emp')
df_dummy = pd.concat([df, gender_dummy, income_level_dummy, employment_status_dummy], axis=1)
df_dummy = df_dummy.drop(columns=['user_id','app_name','country','gender', 'income_level', 'employment_status'])
df_dummy.head()
```
