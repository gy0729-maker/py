```python
from transformers import GPT2LMHeadModel, PreTrainedTokenizerFast

tokenizer = PreTrainedTokenizerFast.from_pretrained("skt/kogpt2-base-v2")
model = GPT2LMHeadModel.from_pretrained("skt/kogpt2-base-v2")
model.eval()

prompt = "작은 마을의 밤, 신비로운 상자가 발견되었다."

input_ids = tokenizer.encode(prompt, return_tensors="pt")

#input_ids : 모델에게 주는 시작 글(입력 문장)
#max_length=80 : 80단어 생성
#temperature=0.7 : 높으면 다양하고 창의적인 내용
#top_k=50 : 후보 단어를 50개로 제한
#top_p=0.9 : 확률 합이 0.9 안에 들어가는 단어만 선택 - 너무 이상한 단어는 안 나오도록
#do_sample=True : 글을 랜덤 방식으로 생성
#no_repeat_ngram_size=2 : 같은 2단어 조합이 반복되지 않도록

output = model.generate(
    input_ids,
    max_length=80,
    temperature=0.7,
    top_k=50,
    top_p=0.9,
    do_sample=True,
    no_repeat_ngram_size=2
)

# 숫자를 텍스트로 변환, 모델 전용 태그는 제외
story = tokenizer.decode(output[0], skip_special_tokens=True)

sentences = story.replace("\n", " ").split(". ")
short_sentences = [s.strip() for s in sentences if 3 < len(s) <= 50]
final_story = ". ".join(short_sentences)

#문장 끝이 마침표로 끝나는지 검사
if not final_story.endswith("."):
    final_story += "."

print(final_story)
```
