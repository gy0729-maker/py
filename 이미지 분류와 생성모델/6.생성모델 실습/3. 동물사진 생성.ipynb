```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/sd-turbo",
    torch_dtype=torch.float32
)

prompt = (
    "a realistic photo of a dog on a table, "
    "natural proportions, detailed fur, "
    "soft studio lighting, simple clean background, "
    "sharp focus, high quality, photorealistic"
)

image = pipe(
    prompt,
    num_inference_steps=8,
    guidance_scale=7.0
).images[0]

image.save("animal.png")
print("이미지 생성 완료: animal.png")
```
