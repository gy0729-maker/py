```python
import torch
from diffusers import StableDiffusionPipeline
import time

model_id = "runwayml/stable-diffusion-v1-5"

print("모델 로딩 중...")
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float32
)

prompts = [
    "a house",
    "a small house",
    "a tree",
    "a mountain",
    "the sea",
    "an apple",
    "a cat",
    "a dog",
    "a flower",
    "the sky"
]



print("이미지 생성 시작...")

for i, prompt in enumerate(prompts, start=1):
    start = time.time()
    print(f"[{i}] 생성 중: {prompt}")

    image = pipe(
        prompt=prompt,
        num_inference_steps=2,
        guidance_scale=0.0,
        width=256,
        height=256
    ).images[0]

    elapsed = time.time() - start

    filename = f"output_{i}_{prompt}.png"
    image.save(filename)

    print(f"저장 완료: {filename} ({elapsed:.1f}초)")

print("전체 생성 완료")
```
