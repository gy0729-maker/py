```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('c:/images/colorball.jpg')
#img = cv2.imread('c:/images/apple.jpg')
src = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.axis('off')
plt.show()
```

```python
import cv2
dst = src.copy()
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
#cv2.HoughCircles(이미지, 검출 방법, 해상도 비율, 최소 거리, 캐니 엣지 임계값, 중심 임계값, 최소 반지름, 최대 반지름)
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 100, param1=250, param2=10, minRadius=80, maxRadius=120)

for i in circles[0]:
    cv2.circle(dst, (int(i[0]), int(i[1])), int(i[2]), (255, 255, 255), 5)
```

```python
plt.imshow(dst)
plt.title('image1')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# img = cv2.imread('c:/images/colorball.jpg')
img = cv2.imread('c:/images/apple.jpg')
src = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.axis('off')
plt.show()
import cv2

dst = src.copy()
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)                   #이미지 사이즈에 따라 반지름의 범위설정이 중요
#cv2.HoughCircles(이미지, 검출 방법, 해상도 비율, 최소 거리, 캐니 엣지 임계값, 중심 임계값, 최소 반지름, 최대 반지름)
# circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 100, param1=250, param2=10, minRadius=80, maxRadius=120)
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 100, param1 = 250, param2 = 10, minRadius = 120, maxRadius = 130)

for i in circles[0]:
    cv2.circle(dst, (int(i[0]), int(i[1])), int(i[2]), (255, 255, 255), 5)
plt.imshow(dst)
plt.title('image1')
plt.axis('off')
plt.show()
```
