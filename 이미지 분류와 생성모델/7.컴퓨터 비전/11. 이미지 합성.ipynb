```python
###########################
import cv2
import matplotlib.pyplot as plt
img1 = cv2.imread("c:/images/drawing.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
###########################
img2= cv2.imread("c:/images/my_hand.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.axis("off")
plt.show()
```

```python
###########################
import numpy as np
# mask: blank image = 같은 크기와 자료형을 가지는 배열을 만들고 255로 채움
mask = np.full_like(img1, 255)
# 합성 대상 좌표 계산(img2의 중앙)
height, width = img2.shape[:2]
center = (width//2, height//2)
```

```python
img2.shape
```

```python
###########################
# seamlessClone
# seam 경계 less 없는 clone 복제, 부드럽게 이미지를 합성하는 함수
normal = cv2.seamlessClone(img1, img2, mask, center, cv2.NORMAL_CLONE)
plt.imshow(normal)
plt.axis("off")
plt.show()
# 꽃은 선명하지만 주변 이미지가 흐려짐
```

```python
###########################
mixed = cv2.seamlessClone(img1, img2, mask, center, cv2.MIXED_CLONE)
plt.imshow(mixed)
plt.axis("off")
plt.show()
# 꽃은 다소 흐리지만 주변 이미지와 잘 매칭이 되게 합성이 됨
```

```python
mixed = cv2.seamlessClone(img1, img2, mask, center, cv2.NORM_L2)
plt.imshow(mixed)
plt.axis("off")
plt.show()
```
