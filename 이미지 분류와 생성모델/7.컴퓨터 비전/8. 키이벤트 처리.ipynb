```python
import cv2

img = cv2.imread('c:/images/winter.jpg')
cv2.namedWindow('img')
cv2.imshow('img', img)

while True: #무한반복
    keycode = cv2.waitKey()
    if keycode == ord('i') or keycode == ord('I'):
        img = ~img #비트반전
        cv2.imshow('img', img)
                   # 27:esc  /  ord=> 키를 고유의 숫자로 변환
    elif keycode == 27 or keycode == ord('q') or keycode == ord('Q'):
        break

cv2.destroyAllWindows()
```
