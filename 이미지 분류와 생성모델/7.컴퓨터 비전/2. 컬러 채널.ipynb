```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('c:/images/rear_garden.png')
print(img.shape)
#세로, 가로, 채널(흑백1/컬러3)
```

```python
(height, width) = img.shape[:2]
center = (width // 2, height // 2)
#convert color 컬러변환 opencv:BGR => matplotlib:RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#이미지 채널 분할
(Blue, Green, Red) = cv2.split(img)
#img 사이즈만큼 0을 만듦
zeros = np.zeros(img.shape[:2], dtype="uint8")
```

```python
#red
img = cv2.cvtColor(cv2.merge([zeros, zeros, Red]), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#green
img = cv2.cvtColor(cv2.merge([zeros, Green, zeros]), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#blue
img = cv2.cvtColor(cv2.merge([Blue, zeros, zeros]), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
BR = cv2.merge([Blue, zeros, Red])
plt.imshow(BR)
```

```python
#분할된 3채널 합치기
BGR = cv2.merge([Blue, Green, Red])
plt.imshow(BGR)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```
