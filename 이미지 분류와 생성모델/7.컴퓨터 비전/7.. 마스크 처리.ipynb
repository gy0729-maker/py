```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('c:/images/rear_garden.png')
(height, width) = img.shape[:2]
center = (width // 2, height // 2)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#mask 생성
#img size의 검은화면 생성
mask = np.zeros(img.shape[:2], dtype='uint8')
#mask에 filter 출력 (200사이즈의 하얀 원 그리기)
cv2.circle(mask, center, 200, (255, 255, 255), -1)
plt.imshow(mask)
```

```python
#이미지에 mask 적용
#bitwise_and 둘 다 1일경우에만 1 (img,mask 모두 1인 원안의 공간만 합침)
masked = cv2.bitwise_and(img, img, mask=mask)
plt.imshow(masked)
```
