```python
#모집단 분산
import numpy as np

point=[40,50,60]
a=np.var(point)

print(a)

#분모 = n = 3
#분산 = 200 / 3 ≈ 66.67
```

```python
#표본 분산 -데이터가 일부 표본일 때 사용, 과소추정 방지를 위해 베셀보정효과
from statistics import variance

point = [40,50,60]
b = variance(point)

print(b)

# 분모 = n - 1 = 2
# 분산 = 200 / 2 = 100
```

```python
#표준편차 == 분산의 제곱근
#모집단 표준편차
import numpy as np

point = [40,50,60]
c = np.std(point)

print(c)

# 머신러닝에서는 데이터가 모집단이 아니라 표본이라는 가정 하에 분산과 표준편차의 과소추정을 방지하고
# 더 안정적인 일반화를 위해 ddof=1을 자주 사용한다.
c1 = np.std(point, ddof=1)
print(c1)
```

```python
#표본 표준편차
from statistics import stdev

point = [40,50,60]
d = stdev(point)

print(d)
```
