```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense
```

```python
df = pd.read_csv('c:/data/test/sales_time.csv')
values = df['sales'].values.reshape(-1,1) #n행 1열로 reshape
values
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)
```

```python
def make_sequence(data, window=3):
    X,y = [],[]         #타입스텝 3
    for i in range(len(data) - window):
        X.append(data[i : i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)
```

```python
X, y = make_sequence(scaled, window=3)
model = Sequential()
model.add(Input(shape=(3,1)))
model.add(LSTM(32))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X,y, epochs=200, verbose=1)
model.save('c:/data/test/sales_rnn_model.keras')
print('모델 저장 완료')
```
