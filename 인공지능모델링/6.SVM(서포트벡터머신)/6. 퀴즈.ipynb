```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
csv_path='c:/data/test/exam_pass.csv'
data = pd.read_csv(csv_path)
data.head()
```

```python
target_col = 'pass'
X = data.drop(columns=[target_col])
y = data[target_col]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
svm = SVC(kernel='linear', C=1)
svm.fit(X_train_s, y_train)
pred = svm.predict(X_test_s)
print('Accuracy:',accuracy_score(y_test, pred))
```

```python
print('Confusion matrix:\n', confusion_matrix(y_test, pred))
```

```python
new_student = pd.DataFrame([{'study_hours':9, 'sleep_hours':6, 'stress':5}])
new_student_s = scaler.transform(new_student)
print('새 학생 예측(0=불합격,1=합격):', svm.predict(new_student_s)[0])
```
