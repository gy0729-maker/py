```python
# exp : exponential(지수함수)
# exp는 입력이 조금만 바뀌어도 출력과 기울기가 즉시 반응하기 때문에 학습 신호를 전달하는 데 유리하다
import math

print(math.exp(1))
print(math.exp(2))
print(math.exp(3))
```

```python
for x in range(4):
    print(
        'x=', x,
        '2**x=', 2**x,
        'exp(x)=', round(math.exp(x), 2)
    )
```

```python
a=80
b=82
ea=math.exp(a)
eb=math.exp(b)
print('원래값의 차:',b-a)
print('exp(80)=', ea)
print('exp(82)=', eb)
print('exp 통과 후 비율비교:', eb/ea, '배')
```

```python
import matplotlib.pyplot as plt

x=[i*0.1 for i in range(0,51)]
y_pow2=[2**xi for xi in x]
y_exp=[math.exp(xi) for xi in x]

plt.figure()
plt.plot(x,y_pow2, label='2**x')
plt.plot(x,y_exp, label='exp(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('value')
plt.title('2**x vs exp(x)')
plt.show()
```
