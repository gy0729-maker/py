```python
# 지수와 로그는 역함수 관계
import math

for n in range(0, 50, 5):
    exp_value = 2 ** n
    log_value = math.log2(exp_value)
    print(f'{n}, 지수:{exp_value}, 로그:{log_value}')
```

```python
import math
n=3
exp_n = 2 ** n
back = math.log2(exp_n)
print('원래 값:', n)
print('지수:', exp_n)
print('다시 로그:', back)
```

```python
p = 0.0001
normal = 1.0
for _ in range(100):
    normal *= p
print(f'{normal}')
```

```python
log_sum = 0.0
for _ in range(100):
    log_sum += math.log(p)
print(f'{log_sum:.2f}')
```

```python
# 지수와 로그 그래프
import math
import matplotlib.pyplot as plt

x = list(range(1,51))
y_small = [0.5**i for i in x]
y_log = [math.log(v) for v in y_small]
```

```python
plt.plot(x, y_small)
plt.title('exp')
plt.show()
#점점 작아져서 그래프가 둔감해짐
```

```python
plt.plot(x, y_log)
plt.title('log')
plt.show()
#로그로 변환하면 변화가 확실히 보임
```
