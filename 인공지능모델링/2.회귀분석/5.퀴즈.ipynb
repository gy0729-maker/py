```python
import pandas as pd

df = pd.DataFrame({
    'hours': [1,2,3,4,5],
    'score': [50,55,65,70,75]
})
df
```

```python
X = df[['hours']] #독립변수는 여러개 넣는 경우가 많아 리스트로 넣는다
y = df['score']

#선형 회귀모델 불러오기
from sklearn.linear_model import LinearRegression
#모델 객체 만들기
model = LinearRegression()
#모델 학습
model.fit(X,y)
#학습된 모델로 예측하기
y_pred = model.predict(X)
#실제값고 예측값 사이에 약간의 오차가 있음
print(y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y)
plt.show()
```

```python
#기울기(계수) 확인
print(model.coef_)
```

```python
#절편 확인
print(model.intercept_)
```

```python
#회귀식 도출됨
# y = 6.5 * X + 43.5
```

```python
#공부시간이 6시간일 때 점수예측
pred = model.predict([[6]])
print(pred)
```
