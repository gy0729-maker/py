```python
import pandas as pd
df = pd.read_csv('c:/data/test/house.csv')
print(df.head())
```

```python
X = df[['area','rooms','age']]
y = df['price']
```

```python
import statsmodels.api as sm

model = sm.OLS(y,X)
result = model.fit()
```

```python
print(result.summary())
```
