```python
#입력값
age = 30
experience = 5
accident = 1

#계산의 편의를 위해 나이를 10으로 나눔(스케일링)
x1 = age/10
x2 = experience
x3 = accident

y_true = 100 #실제보험료

#초기 가중치(weight)
w1 = 1.0
w2 = -1.0
w3 = 5.0
#편향/절편(bias) : 기본보험료
b = 50.0

learning_rate = 0.01  #학습률
epochs = 10  #전체 학습 반복횟수(순전파,역전파 한사이클)
errors = []  #오차 저장용 리스트

print('초기 가중치:', w1, w2, w3, b)
print('-'*40)

#학습
for epoch in range(1, epochs +1):
    #순전파 (예측값, 오차 계산)
    y_pred = x1 * w1 + x2 * w2 + x3 * w3 + b
    error = y_pred - y_true
    errors.append(error)

    print(f'[에폭 {epoch}]')
    print('예측값:', y_pred)
    print('오차(예측-실제):', error)
    #역전파 (가중치 업데이트)
    w1 = w1 - learning_rate * error * x1
    w2 = w2 - learning_rate * error * x2
    w3 = w3 - learning_rate * error * x3
    b = b - learning_rate * error

    print(f'수정된 가중치: {w1:.2f}, {w2:.2f}, {w3:.2f}, {b:.2f}')
    print('-'*40)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, epochs + 1), errors, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Error (Prediction - True)')
plt.title('Error Decreasing Over Epochs')
plt.grid(True)
plt.show()
```
