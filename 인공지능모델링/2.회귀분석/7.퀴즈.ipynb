```python
import pandas as pd
import statsmodels.api as sm

df = pd.DataFrame({
    'hours': [1,2,3,4,5],
    'sleep': [6,7,6,8,7],
    'score': [50,58,65,72,78]
})
df
```

```python
df.plot(x='hours', y='score')
```

```python
X = df[['hours','sleep']]
y = df['score']
```

```python
#OlS 모델 생성/학습
model = sm.OLS(y,X)
result = model.fit()
result.summary()
```

```python
print(result.params)
```

```python
print(result.pvalues)
```

```python
print(result.rsquared)
```

```python
R = result.rsquared**0.5
print(R)
```

```python
y_pred = result.predict(X)
print(y_pred)
```

```python
# hours의 p-value가 0.05보다 작다면, hours는 score에 유의한 영향을 준다고 해석한다.
# R-squared 값이 0.9라면, 이 모델은 score 변동의 약 90%를 설명한다.
```
