```python
from sklearn.datasets import make_regression

X,y,w = make_regression(n_samples=200, n_features=1, bias=100, noise=10, random_state=1, coef=True)
print(X[:5])
print(y[:5])
print(w)
```

```python
import pandas as pd
df = pd.DataFrame({'x':X.flatten(), 'y':y})
df
```

```python
import statsmodels.api as sm

X = df[['x']]
y = df[['y']]
```

```python
model = sm.OLS(y,X)
result = model.fit()
result.summary()
```

```python
#새로운 값 예측
result.predict([-2,-1,0,1,2])
```

```python
#가중치(기울기)
result.params
```

```python
#잔차(실제값과 예측값의 차이)
result.resid
```

```python
import matplotlib.pyplot as plt
#잔차 벡터 그래프
result.resid.plot(style='o')
plt.xlabel('index')
plt.ylabel('Residual')
```
