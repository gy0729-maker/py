```python
from sklearn.datasets import make_regression
#회귀 분석을 위한 가상 데이터셋과 실제 회귀 계수(가중치)를 생성하는 코드
X,y,w = make_regression(n_samples=50, n_features=1, bias=100, noise=10, random_state=0, coef=True)
print(X[:5].flatten()) #독립변수X는 2차원 배열
print(y[:5])
print(w) #기울기(계수)
```

```python
import numpy as np
xx = np.linspace(-3, 3, 100) #-3~3, 100등분
y0 = w * xx +100
```

```python
import matplotlib.pyplot as plt
plt.plot(xx, y0, 'r-')
plt.scatter(X, y, s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('regression')
```

```python
from scipy import stats
stats.linregress(X.flatten(), y.flatten())
```
