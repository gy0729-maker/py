```python
import pandas as pd
df = pd.read_csv('c:/data/test/score.csv')
print(df)
```

```python
X = df[['study_time','attendance','assignment']]
y = df['score']
```

```python
# LinearRegression 모델 생성/학습
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

```python
# OLS 모델 생성/학습
import statsmodels.api as sm
result = sm.OLS(y,X).fit()
result.summary()
```

```python
# OLS는 회귀계수 coef 뿐 아니라 통계적 유의성 p-value도 확인할 수 있다.
# OLS는 주로 모형 분석/해석에 사용된다.
# LinearRegression은 주로 예측에 사용된다
```
