```python
import pandas as pd
data = pd.read_csv('c:/data/insurance/insurance.csv')
data.head()
```

```python
data2 = pd.read_csv('c:/data/insurance/insurance2.csv')
data2.head()
```

```python
data2.corr()['charges'].sort_values()
```

```python
X = data2[data2.columns[:-1]]
y = data2['charges']
```

```python
import statsmodels.api as sm
model = sm.OLS(y,X)
result = model.fit()
result.summary()
```

```python
#P-value 0.05이상인 'sex'제거
X = X[['age', 'bmi', 'children', 'smoker', 'southwest', 'southeast',
       'northeast', 'northwest']]
model = sm.OLS(y,X)
result = model.fit()
result.summary()
```
