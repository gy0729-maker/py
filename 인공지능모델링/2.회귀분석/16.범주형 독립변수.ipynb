```python
import pandas as pd
df = pd.read_csv('c:/data/degree/data.csv')
df
```

```python
df.boxplot('value','month')
```

```python
import statsmodels.api as sm
model = sm.OLS.from_formula('value~month', df)
            # 수식사용('종속변수 ~ 독립변수')
results = model.fit()
results.summary()
```

```python
import statsmodels.api as sm
model = sm.OLS.from_formula('value ~ C(month)+0', df)
                    #범주형 변수 = C(변수)=>카테고리(범주형) +0:1월부터시작
results = model.fit()
results.summary()
```

```python
results.summary()
import statsmodels.api as sm

model = sm.OLS.from_formula('value ~ C(month)', df)
                        #범주형 변수 = C(변수)=>카테고리(범주형) +0없으면:2월부터시작
results = model.fit()
results.summary()
```
