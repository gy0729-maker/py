```python
import pandas as pd

df=pd.DataFrame({
    'age': [20,25,30],
    'score': [80,90,70],
    'gender': ['M','F','M']
})

df['pass'] = df['score']>=80
print(df)
```

```python
# 컬럼1개==시리즈       시리즈.apply(함수)
df['age_group'] = df['age'].apply(lambda x:'audult' if x>=25 else 'young')
print(df)                #lamda input:output 참일때             거짓일때 
```

```python
df['adj_score'] = df['score']+5
print(df)
```

```python
df['male'] = df['gender'] == 'M'
print(df)
```

```python
df['gender_len'] = df['gender'].str.len()
print(df)
```

```python
df['grade'] = df['score'].apply(lambda x:'A' if x>=90 else 'B')
print(df)
```

```python
df['score_ratio'] = df['score']/100
print(df)
```
