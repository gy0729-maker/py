```python
from patsy import demo_data
import pandas as pd

df=pd.DataFrame(demo_data('x1','x2','x3','x4','x5'))
df
```

```python
import matplotlib.pyplot as plt
df.boxplot()
plt.show()
```

```python
#평균 0, 표준편차 1이 되도록 스케일링
###딥러닝에 꼭 필요한 StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df2=scaler.fit_transform(df)
df3=pd.DataFrame(df2, columns=df.columns)
df3
```

```python
df3.boxplot()
plt.show()
```

```python
import numpy as np
X = np.arange(7).reshape(7,1) #7행,1열의 2차원으로 변환
 #arange(7)=0~6
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler =  StandardScaler()
X2 = scaler.fit_transform(X)
X2
```

```python
X2 = np.vstack([X, [[100]]]) #배열을 세로로 쌓는 함수
X2
```

```python
import matplotlib.pyplot as plt
pd.DataFrame(X2).boxplot()
plt.show()
```

```python
scaler.fit_transform(X2)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
X3 = scaler.fit_transform(X2)
X3
```

```python
import matplotlib.pyplot as plt
pd.DataFrame(X3).boxplot()
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
X4 = scaler.fit_transform(X3)
X4
```
