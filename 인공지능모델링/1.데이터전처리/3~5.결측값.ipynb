```python
#4. 결측값 개수
import pandas as pd
data = {
    '이름': ['철수','영희',None,'지영','수진'],
    '나이': [25, None, 19, 22, None],
    '점수': [82, 91, None, 74, 88],
    '성별': ['남성','여성','남성',None,'여성']
}
df = pd.DataFrame(data)
print(df)
print(df.isnull()) #결측값이면 True
print(df.isnull().sum()) #컬럼별 결측값수 계산
```

```python
data = {
    '이름': ['철수','영희','민수','지영','수진',None],
    '나이': [25, 30, None, 22, 28, 30],
    '점수': [82, 91, 75, None, 88, 75],
    '성별': ['남성','여성','남성','여성',None,'여성']
}
df=pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data=data, x='점수', bins=5)
```

```python
#평균값으로 결측값 처리
df['나이']=df['나이'].fillna(df['나이'].mean())

#중위수로 결측값 처리
df['점수']=df['점수'].fillna(df['점수'].median())

#최빈값으로 결측값 처리
df['성별']=df['성별'].fillna(df['성별'].mode()[0])

#고정값으로 결측값 처리
df['이름']=df['이름'].fillna('미상')
df
```
