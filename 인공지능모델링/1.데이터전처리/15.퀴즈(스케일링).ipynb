```python
import pandas as pd

df = pd.DataFrame({
    'height': [150, 160, 170],
    'weight': [50, 60, 70]
})
df
```

```python
import matplotlib.pyplot as plt
df.boxplot()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df[['height_std']] = scaler.fit_transform(df[['height']])
print(df)
```

```python
df[['h_std','w_std']] = scaler.fit_transform(df[['height','weight']])
print(df)
```

```python
print(df['height_std'].mean(), df['height_std'].std())
```

```python
import matplotlib.pyplot as plt

df[['h_std','w_std']].boxplot()
plt.show()
```
