```python
import pandas as pd

data = {
    '이름': ['철수','영희','민수','지영','수진'],
    '성별': ['남성','여성','남성','여성','여성'],
    '등급': ['B','A','C','C','B'],
    '지역': ['서울','부산','서울','대구','부산']
}
df=pd.DataFrame(data)
df
```

```python
gender_dummy = pd.get_dummies(df['성별'], prefix='성별')
gender_dummy
```

```python
grade_dummy = pd.get_dummies(df['등급'], prefix='등급')
region_dummy = pd.get_dummies(df['지역'], prefix='지역')
#기존데이터 + 더미변수 결합
df_dummy = pd.concat([df, gender_dummy, grade_dummy, region_dummy], axis=1) #axis=1 가로방향
df_dummy
```

```python
df_dummy = df_dummy.drop(columns=['성별','등급','지역'])
df_dummy
```
