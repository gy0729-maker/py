```python
import pandas as pd

data={
    '이름':['철수','영희','민수','지영','수진'],
    '점수':[82, 91, 150, 74, 88],
    '나이':[25, 30, 22, 120, 28]
}
df = pd.DataFrame(data)
df
```

```python
#IQR 방식 이상치 처리
Q1=df['점수'].quantile(0.25)
Q3=df['점수'].quantile(0.75)
IQR = Q3 - Q1
#박스길이의 위아래 1.5배를 넘어가면 이상치로 처리하여 제거
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df= df[(df['점수']>=lower) & (df['점수']<=upper)]
df
```

```python
df=df[df['나이']<=100]
df
```
