```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'age': [20, np.nan, 30],
    'score': [80, np.nan, 90]
})
df
```

```python
print(df.isnull())  #isnull=isna
```

```python
print(df.isnull().sum())
```

```python
df2 = df.dropna()
print(df2)
```

```python
df2 = df.fillna(0)
print(df2)
```

```python
df2 = df.copy()
df2['score'] = df['score'].fillna(0)
print(df2)
```

```python
df2 = df.copy()
avg = df2['age'].mean()
df2['age'] = df2['age'].fillna(avg)
print(df2)
```

```python
print(df['age'].isnull())
```
