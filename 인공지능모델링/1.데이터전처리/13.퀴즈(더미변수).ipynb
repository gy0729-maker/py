```python
import pandas as pd

df = pd.DataFrame({
    'gender': ['M','F','F','M'],
    'city': ['Seoul','Busan','Seoul','Incheon']
})
df
```

```python
d1 = pd.get_dummies(df['gender'])
print(d1)
```

```python
d2 = pd.get_dummies(df[['gender','city']])
print(d2)
```

```python
dummy = pd.get_dummies(df['gender'])
df2 = pd.concat([df, dummy], axis=1)
print(df2)
```

```python
d3 = pd.get_dummies(df['gender'], drop_first=True)
print(d3)
```

```python
d4 = pd.get_dummies(df['city'], prefix='city')
print(d4)
```
