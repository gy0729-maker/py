```python
import pandas as pd

data = {
    '이름': ['철수','영희','민수','지영','수진'],
    '나이': [25, 30, 19, 22, 28],
    '점수': [82, 91, 67, 74, 88],
    '출석률': [0.95, 0.88, 0.76, 0.92, 0.85]
}
df=pd.DataFrame(data)
df
```

```python
#점수 구간 파생변수
df['등급'] = 'C' #새로운 변수추가
df.loc[df['점수']>=80, '등급']='B'
    #loc[행, 열]
df.loc[df['점수']>=90, '등급']='A'
df
```

```python
df['성인여부']=df['나이']>=20
df
```

```python
df['출석통과']=df['출석률']>=0.9
df
```

```python
# 최종합격 여부 = (점수) + (출석률)
df['최종합격']=(df['점수']>=80) & (df['출석률']>=0.9)
df
```

```python
df['보정점수']=df['점수']+(df['출석률']*10)
df
```
