```python
import pandas as pd

df = pd.DataFrame({
    'score': [50,55,60,65,200],
    'age': [20,22,23,24,180]
})
df
```

```python
import matplotlib.pyplot as plp
df.boxplot()
```

```python
print(df.describe())
```

```python
Q1 = df['score'].quantile(0.25)
Q3 = df['score'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

print(lower)
print(upper)
```

```python
outlier = df[(df['score']<lower) | (df['score']>upper)]
print(outlier)              #   or
```

```python
df2 = df[(df['score']>=lower) & (df['score']<=upper)]
print(df2)
```

```python
df['score_fixed']=df['score'].apply(
    lambda x: upper if x>upper else x
)
print(df)
```
