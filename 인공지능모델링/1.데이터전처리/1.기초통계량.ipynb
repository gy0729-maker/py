```python
import numpy as np
from scipy import stats

data = np.array([4,5,1,2,7,2,6,9,3])
```

```python
#평균
a=np.mean(data)
print(a)
print(f'{a:.2f}')
```

```python
#중위수
b=np.median(data)
print(b)
```

```python
#최빈값
c=stats.mode(data)
print(c[0])
```

```python
from statistics import variance,stdev
points=np.array([20,80,90,95,87,89,95,99,97,100,60,70,77,88,89,89,90])
```

```python
#분산
a=variance(points)
print(a)
```

```python
#표준편차
b=np.std(points)
print(b)
```

```python
#범위
c=np.max(points) - np.min(points)
print(c)
```

```python
#최대,최소값
a=np.max(points)
b=np.min(points)
c=a-b
print(a)
print(b)
print(c)
```

```python
#백분위
for val in [20,80,100]:
    d=np.percentile(points,val)
    print(str(val)+'%', d)
```

```python
#사분위수
a,b,c=np.percentile(points,[25,50,75])
print(a)
print(b)
print(c)
print(c-a) #IQR(InterQuartile Range)
```
