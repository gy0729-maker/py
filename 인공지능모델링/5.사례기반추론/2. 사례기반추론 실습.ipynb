```python
## 지도학습
from sklearn.neighbors import KNeighborsClassifier

X = [[0],[1],[2],[3]]
y = [0,0,1,1]
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X,y)
# 1.1의 분류는? ->0으로 분류됨
print(knn.predict([[1.1]]))
```

```python
# 0.9는 0일 확률 66%, 1일 확률 33% ->0으로 분류됨
print(knn.predict_proba([[0.9]]))
print(knn.predict([[0.9]]))
```

```python
## 비지도학습
from sklearn.neighbors import NearestNeighbors

samples = [[0,0,0],[0,0.5,0],[1,1,0.2]]
nn = NearestNeighbors(n_neighbors=1)
nn.fit(samples)
# 새로운 데이터 1,1,1에 가장 가까운 데이터는? -> 세번째가 가장 가깝고 거리는 0.8
print(nn.kneighbors([[1.,1.,1.]]))
```

```python
## 최근접 이웃 회귀모형
from sklearn.neighbors import KNeighborsRegressor

X= [[40],[45],[60],[70]]  #평방미터
y = [20,22,30,35]  #월세
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X,y)
# 65평방미터일 때의 월세는? ->32.5
print(neigh.predict([[65]]))
print(neigh.predict(X))
```
