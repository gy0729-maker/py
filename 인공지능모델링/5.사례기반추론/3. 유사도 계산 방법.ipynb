```python
import matplotlib.pyplot as plt
from fontTools.mtiLib import parseLookupRecords

X = [10,25,47,68,89]
y = [20,40,55,65,75]
plt.figure(figsize=(5,5))
plt.scatter(X,y)
mypoint = (80,5)
plt.scatter(mypoint[0],mypoint[1],s=100,c='r')
```

```python
from math import sqrt
import numpy as np
def distance(p,m):
    items=[]
    for i in range(len(p-1)):
        d = sqrt((m[0] - p[i][0])**2 + (m[1] - p[i][1])**2)
        items.append(d)
    return items

points = np.array(list(zip(X,y)))
r = distance(points, mypoint)
print(r)
m=min(r)
i=np.argmin(r)
print('가장 가까운 포인트의 인덱스:',i)
print('가장 가까운 포인트는?', points[i])
```

```python
#사이파이 패키지에 내장된 유클리드 거리 함수 =>점과 점사이의 직선거리
from scipy.spatial import distance
results=[]
for point in points:
    dist = distance.euclidean(point,mypoint)
    results.append(dist)
results
```

```python
#맨해튼 거리 구하기 =>점까지 가는 가로길이+세로길이
def distance(p,m):
    items=[]
    for i in range(len(p-1)):
        d = abs(m[0] - p[i][0]) + abs(m[1] - p[i][1])
        items.append(d)
    return items

r = distance(points, mypoint)
print(r)
m = min(r)
i=np.argmin(r)
print('가장 가까운 포인트의 인덱스:',i)
print('가장 가까운 포인트는?',points[i])
```

```python
#사이파이 패키지에 내장된 맨해튼 거리를 구하는 함수
from scipy.spatial import distance
results=[]
for point in points:
    dist = distance.cityblock(point,mypoint)
    results.append(dist)
results
```
