```python
from sklearn.datasets import make_blobs #랜덤데이터를 만드는 모듈
from sklearn.cluster import KMeans
#인위적인 2차원 데이터셋
X, y = make_blobs(random_state=1) #cluster_std=1
#군집 모형 생성
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
lbl = model.labels_
lbl
```

```python
#예측한 값
pred = model.predict(X)
print(pred)
```

```python
import mglearn
import matplotlib.pyplot as plt
#3개의 클러스터를 산점도로 표현 - X1,  X2,  y
mglearn.discrete_scatter(X[:,0],X[:,1],lbl,markers='o')
#centers
mglearn.discrete_scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],[0,1,2],markers='^',c='r',s=20)
plt.show()
```

```python
from matplotlib import pyplot as plt
fig, axes = plt.subplots(1,4,figsize=(20,5))
idx=0
for k in [2,3,4,5]:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    lbls = kmeans.labels_
    mglearn.discrete_scatter(X[:,0], X[:,1], lbls, ax=axes[idx])
    idx+=1

plt.show()
```
