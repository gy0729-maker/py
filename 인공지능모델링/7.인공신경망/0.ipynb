```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.utils import to_categorical

df = pd.read_csv("c:/data/test/product_quality.csv")
df
```

```python
X = df.drop("label", axis=1).values
y = df["label"].values
X
```

```python
#yк°’ мқёмҪ”л”©
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_categorical = to_categorical(y_encoded)
print(y_encoded)
y_categorical
```

```python
# н•ҷмҠөмҡ©:кІҖмҰқмҡ© лҚ°мқҙн„°м…Ӣмқ„ 8:2лЎң лӮҳлҲ•лӢҲлӢӨ.
X_train, X_test, y_train, y_test = train_test_split(
    X, y_categorical, test_size=0.2, random_state=42, stratify=y_categorical
)
```

```python
#нҡҢк·Җл¶„м„қм—” мҠӨмјҖмқјл§Ғ н•„мҲҳ!!
scaler = StandardScaler()
#н•ҷмҠөмҡ© лҚ°мқҙн„°лҘј кё°мӨҖмңјлЎң лҚ°мқҙн„°мқҳ нҢЁн„ҙмқ„ л¶„м„қн•ҳкі  ліҖнҷҳн•©лӢҲлӢӨ.
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#мӢ кІҪл§қ лӘЁлҚё мғқм„ұ
model = Sequential([
    Input(shape=(X_train.shape[1],)),  # мІ« л Ҳмқҙм–ҙлЎң Input мӮ¬мҡ©
    # нҷңм„ұнҷ” н•ЁмҲҳлҘј reluлЎң м„Өм •н•©лӢҲлӢӨ. hiddenм—җм„ң л§Һмқҙ м“°мһ„ (мқҢмҲҳ=>0, м–‘мҲҳлҠ” к·ёлҢҖлЎң м¶ңл Ҙ)
    Dense(8, activation='relu'),
    Dense(8, activation='relu'),
    # нҷңм„ұнҷ” н•ЁмҲҳ : sigmoid; 0.0~1.0(мқҙм§„л¶„лҘҳ) / softmax; нҷ•лҘ к°’(лӢӨл¶„лҘҳ)
    Dense(y_categorical.shape[1], activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# validation лҚ°мқҙн„°м…Ӣмқ„ 20%лЎң м„Өм •н•ҳкі  epochмқ„ 50мңјлЎң м„Өм •н•©лӢҲлӢӨ.
history = model.fit(X_train, y_train, validation_split=0.2, epochs=50, batch_size=4, verbose=1)

loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"н…ҢмҠӨнҠё м •нҷ•лҸ„: {accuracy:.4f}")
```
