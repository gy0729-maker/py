```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

df = pd.read_csv('c:/data/test/student_score.csv')
df
```

```python
X = df[['study_hours','attendance_rate','assignment_score']].values
y = df[['final_score']].values
```

```python
print(X.shape)
```

```python
model = Sequential()
model.add(Input(shape=(3,)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X,y,epochs=500,verbose=0)
```

```python
model.summary()
```

```python
model.save('c:/data/test/student_model.keras')
print('모델 학습 및 저장완료')
```

```python
import numpy as np

x1 = np.array([[6, 90, 75]],)
model.predict(x1)
```
