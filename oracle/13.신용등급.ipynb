```python
import oracledb

dsn =  oracledb.makedsn('localhost',1521,service_name='xe')

conn = oracledb.connect(
    user='python',
    password='1234',
    dsn=dsn
)
print('DB 접속 성공')
```

```python
cursor = conn.cursor()
cursor.execute('select * from credit_risk order by customer_id')
rows = cursor.fetchall()
```

```python
import pandas as pd

df=pd.DataFrame(
    rows,
    columns=[
        'customer_id',
        'age',
        'income_monthly',
        'debt_ratio',
        'credit_score',
        'late_payments',
        'risk_grade',
    ]
)
print(df.head())
```

```python
backup_path = 'credit_risk.csv'
df.to_csv(backup_path, index=False)
print('CSV 백업 저장 완료:', backup_path)
```
