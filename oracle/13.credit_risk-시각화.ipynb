```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('credit_risk.csv')
print(df.head())
```

```python
sns.set_theme(style='whitegrid')
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
grade_map = {0:'저위험(0)', 1:'중위험(1)', 2:'고위험(2)'}
df['grade_name'] = df['risk_grade'].map(grade_map)
print(df[['risk_grade','grade_name']].head())
```

```python
plt.figure()
sns.countplot(x='grade_name', data=df, order=['저위험(0)', '중위험(1)',  '고위험(2)'])
plt.title('신용 위험 등급 분포(3-class)')
plt.xlabel('risk_grade')
plt.ylabel('count')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='grade_name',y='credit_score',data=df,order=['저위험(0)','중위험(1)','고위험(2)'])
plt.title('등급별 신용점수 분포(Boxplot)')
plt.xlabel('risk_grade')
plt.ylabel('credit_score')
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    x='debt_ratio',
    y='credit_score',
    hue='grade_name',
    data=df,
    alpha=0.75
)
plt.title('부채비율 vs 신용점수 (등급색상)')
plt.xlabel('debt_ratio(%)')
plt.ylabel('credit_score')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.scatterplot(x='age', y='credit_score', hue='grade_name', data=df, alpha=0.75, legend=False)
plt.title('Age vs Credit Score')


plt.subplot(2,2,2)
sns.scatterplot(x='income_monthly', y='credit_score', hue='grade_name', data=df, alpha=0.75, legend=False)
plt.title('Income vs Credit Score')

plt.subplot(2,2,3)
sns.scatterplot(x='debt_ratio', y='credit_score', hue='grade_name', data=df, alpha=0.75, legend=False)
plt.title('Debt Ratio vs Credit Score')

plt.subplot(2,2,4)
sns.scatterplot(x='late_payments', y='credit_score', hue='grade_name', data=df, alpha=0.75, legend=False)
plt.title('Late Payments vs Credit Score')

plt.show()
```
