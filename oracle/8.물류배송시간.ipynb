```python
#전체데이터 조회
import oracledb
conn = oracledb.connect('python/1234@localhost:1521/XE')

cursor=conn.cursor()
sql = 'select * from delivery order by order_id'
cursor.execute(sql)
rows = cursor.fetchall()

for row in rows[:5]:
    print(row)
```

```python
#선택데이터 조회
sql = '''
select order_id, delivery_time
from delivery
where delivery_time >= :time   --:변수=>매개변수
order by delivery_time
'''

cursor.execute(sql, {'time':100})
for row in cursor.fetchall():
    print(row)
```

```python
#레코드 추가

sql='''
insert into delivery
(order_id, distance_km, package_weight, traffic_level, weather_score, delivery_time)
values(:1, :2, :3, :4, :5, :6)
'''

cursor.execute(sql, (51, 24.0, 6.5, 5.0, 4.0, 98.0))
conn.commit()
print('저장되었습니다.')

sql = 'select * from delivery where order_id = :id'
cursor.execute(sql, {'id': 51})
print(cursor.fetchone())
```

```python
#레코드 수정
        #update 테이블 set 필드 = 변경할값 :변수    where   조건
sql = 'update delivery set delivery_time = :time where order_id = :id'
cursor.execute(sql, {'time': 101.0, 'id': 51})
conn.commit()

sql = 'select * from delivery where order_id = :id'
cursor.execute(sql, {'id': 51})
print(cursor.fetchone())
```

```python
#레코드 삭제

sql='delete from delivery where order_id = :id'
cursor.execute(sql, {'id':51})
conn. commit()

sql='select * from delivery where order_id = :id'
cursor.execute(sql, {'id':51})
print(cursor.fetchone())

cursor.close()
conn.close()
```
