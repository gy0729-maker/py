```python
#물류배송시간 테이블 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

df=pd.read_csv('delivery.csv')
print(df.head())
```

```python
#히스토그램(도수분포)

sns.set_theme(style='whitegrid')
font_path = 'c:/windows/fonts/HanSantteutDotum-Regular.TTF'
font_prop = fm.FontProperties(fname=font_path)

plt.rcParams['font.family'] = font_prop.get_name()
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.hist(df['DELIVERY_TIME'], bins=10)
plt.title('배송 시간 분포')
plt.xlabel('delivery_time(분)')
plt.ylabel('count')
plt.show()
```

```python
#산점도

plt.figure()
sns.scatterplot(
    x='DISTANCE_KM',
    y='DELIVERY_TIME',
    data=df
)
plt.title('배송거리 vs 배송시간')
plt.xlabel('distance_km')
plt.ylabel('delivery_time')
plt.show()
```

```python
#산점도+회귀선

plt.figure()
sns.regplot(
    x='DISTANCE_KM',
    y='DELIVERY_TIME',
    data=df,
    scatter_kws={'alpha':0.8}
)
plt.title('배송거리 vs 배송시간(회귀선)')
plt.xlabel('distance_km')
plt.ylabel('delivery_time')
plt.show()
```

```python
#박스플롯

plt.figure()
sns.boxplot(
    x='TRAFFIC_LEVEL',
    y='DELIVERY_TIME',
    data=df
)
plt.title('교통 혼잡도별 배송 시간 분포')
plt.xlabel('traffic_level')
plt.ylabel('delivery_time')
plt.show()
```

```python
#히트맵

num_cols = [
    'DISTANCE_KM',
    'PACKAGE_WEIGHT',
    'TRAFFIC_LEVEL',
    'WEATHER_SCORE',
    'DELIVERY_TIME'
]

corr = df[num_cols].corr() #상관계수 계산 함수
print(corr)

plt.figure()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='Purples')
plt.title('변수 간 상관관계 히트맵')
plt.tight_layout()
plt.show()
```
