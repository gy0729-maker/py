```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('student_performance.csv')
print(df)
```

```python
sns.set_theme(style='whitegrid')
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

lvl_map = {0:'미흡(0)', 1:'보통(1)', 2:'우수(2)'}
#항목 추가
df['lvl_name'] = df['performance_level'].map(lvl_map) #.map() 모든 샘플 반복처리

plt.figure()
sns.countplot(x='lvl_name', data=df, order=['미흡(0)','보통(1)','우수(2)'])
plt.title('성취도 분포(3-class)')
plt.xlabel('performance_level')
plt.ylabel('count')
plt.show()

print(df)
```

```python
plt.figure()
sns.boxplot(x='lvl_name', y='study_hours', data=df, order=['미흡(0)','보통(1)','우수(2)'])
#sns.boxplot(x='performance_level', y='study_hours', data=df)

plt.title('성취도별 공부시간 분포')
plt.xlabel('performance_level')
plt.ylabel('study_hours')
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    x='study_hours',
    y='assignment_score',
    hue='lvl_name',
    data=df,
    alpha=0.75
)
plt.title('공부시간 vs 과제점수 (성취도 색상)')
plt.xlabel('study_hours')
plt.ylabel('assignment_score')
plt.show()
```
