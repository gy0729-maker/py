```python
import oracledb
import pandas as pd
                        #호스트    포트   데이터베이스
dsn = oracledb.makedsn('localhost',1521,service_name='XE')
#conn = oracledb.connect('python/1234@localhost:1521/XE')

conn = oracledb.connect(
    user='python',
    password='1234',
    dsn=dsn
)
print('DB 접속 성공')
```

```python
cursor = conn.cursor() #sql실행기능
sql = 'select * from student_performance order by student_id'
cursor.execute(sql)
rows = cursor.fetchall()

print('전체 데이터 일부(앞 5개):')
for row in rows[:5]:
    print(row)
```

```python
sql = '''
select performance_level, count(*)
from student_performance
group by performance_level
order by performance_level
'''
cursor.execute(sql)

print('\n클래스 분포:')
for row in cursor.fetchall():
    print(row)
```

```python
sql = '''
select student_id, study_hours, attendance_rate, assignment_score, sleep_hours, performance_level
from student_performance
where performance_level = :lvl
order by student_id
'''
cursor.execute(sql, {'lvl': 2})

good_rows = cursor.fetchall()

print('\n우수(2) 일부(앞 5개:)')
for row in good_rows[:5]:
    print(row)
```

```python
sql='''
select student_id, attendance_rate, assignment_score, study_hours, performance_level
from student_performance
where attendance_rate < :att
and assignment_score < :assign
and study_hours >= :study
order by student_id
'''
cursor.execute(sql, {'att':80, 'assign':65, 'study':5})

warn_rows = cursor.fetchall()
print('\n주의 학생:')
for row in warn_rows:
    print(row)
```

```python
cursor.execute('select * from student_performance order by student_id')
rows = cursor.fetchall()

cursor.close()
conn.close()
print('\nDB 종료')
```

```python
df = pd.DataFrame(
    rows,
    columns=[
        'student_id',
        'study_hours',
        'attendance_rate',
        'assignment_score',
        'sleep_hours',
        'performance_level',
    ]
)
print(df.head())

backup_path = 'student_performance.csv'
df.to_csv(backup_path, index=False)
print('CSV 저장 완료:', backup_path)
```
