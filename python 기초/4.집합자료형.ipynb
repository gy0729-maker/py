```python
scores = [80,90,70,65,100]
total = 0

for score in scores:
    total+=score
print(f'총점은 {total}점입니다.')
```

```python
total=0  #range(start,stop,step)
for i in range(2,101,2):
    total += i

print(f'1~100 짝수 합: {total}')
```

```python
scores = {'김철수': 85, '이영희': 45, '박민수': 60, '최수지': 72, '홍길동': 58}
print('--- 합격자 명단 ---')
for name, score in scores.items():
    if score>=60:
        print(f'합격자: {name}')
```

```python
n = int(input('숫자를 입력하세요 : '))
sum_value=0
i=1

while i <= n:
    sum_value=sum_value+i
    i=i+1

print(f'1부터 {n}까지의 합계는 {sum_value}입니다.')
```
