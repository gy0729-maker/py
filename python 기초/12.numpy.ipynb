```python
import numpy as np

arr1 = np.array([[70,80],[195,185]])
arr2 = np.array([[50,60],[70,80]])
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2) #실수몫
print(arr1 // arr2) #정수몫-소수점이하 버림
print(np.round(arr1/arr2,2)) #넘파이 반올림(값, 소수이하자리수)

print(arr2.shape)
```
