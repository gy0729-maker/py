```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,10,1)
for a in [2, -2]:
    plt.plot(x, a*x, label = 'y={0}x'.format(a))
plt.grid()
plt.legend(fontsize=15, loc=8)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-10,10)
b=np.arange(-5,5,2)
for i in b:
    plt.plot(x, x+i, label=f'y=x+{i}')
plt.grid()
plt.legend(fontsize=13)
plt.show()
```

```python
#relu함수

x = np.arange(-10,11)
y = np.zeros(len(x))
for i in range(len(x)):
    if x[i] >= 0:
        y[i] = x[i]

plt.plot(x,y)
plt.grid()
```

```python
def saving(money, years, rate):
    result = money*(1+rate/100)**years
    return result
# print(saving(10000,1,0.03))

money =10000
years = np.arange(0,11)
rate = [3,5,7]
total_money = np.zeros((len(rate), len(years)))

for i in range(len(rate)):
    for j in range(len(years)):
        total_money[i,j] = saving(money, years[j], rate[i])
    plt.plot(years, total_money[i,:], label='{}%'.format(rate[i]))

plt.legend(fontsize=14)
plt.grid()
plt.xlabel('Years', fontsize=14)
plt.ylabel('Total money', fontsize=14)
plt.xticks(range(0,11,1))
plt.show()
```
