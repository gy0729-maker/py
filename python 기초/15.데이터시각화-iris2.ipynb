```python
import pandas as pd
df=pd.read_csv('c:/data/iris/iris.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
colormap=np.array(['red','lime','black'])
cols=df.columns[0:4]
X=df[cols]
y=df['target']
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False  
y_num, y_text = pd.factorize(y)  
# 번호값(0/1/2),텍스트(setosa,..) 

plt.subplot(1,2,1)
plt.scatter(X['sepal length (cm)'], X['sepal width (cm)'], c=y_num, s=40)
plt.title('꽃받침')

plt.subplot(1,2,2)
plt.scatter(X['petal length (cm)'], X['petal width (cm)'], c=y_num, s=40)
plt.title('꽃잎')
plt.show()
```
