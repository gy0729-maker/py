```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,15,30]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('점수 변화')
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('학생')
plt.ylabel('점수')
plt.show()
```

```python
names=['철수','영희','민수']
scores=[80,95,70]
plt.bar(names, scores)
plt.show()
```

```python
plt.plot(x,y,color='red')
plt.show()
```

```python
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y)
plt.savefig('result.png')
plt.show()
```
