```python
def abs_sign(a):
    if a>=0:
        return a
    else:
        return -a
    
print(abs_sign(5))
print(abs_sign(-3))
```

```python
import math

def abs_square(a):
    b=a*a
    return math.sqrt(b)

print(abs_square(5))
print(abs_square(-3))
```

```python
def mysum(n):
    s=0
    for i in range(1, n+1):
        s=s+i
    return s

print(mysum(10))
print(mysum(100))
print(mysum(1000000))
```

```python
def sum_n(n):
    return n*(n+1)//2

print(sum_n(10))
print(sum_n(100))
print(sum_n(10000000000000000000))
```

```python
def find_max(a):
    n=len(a)
    max_v=a[0]
    for i in range(1,n):
        if a[i] > max_v:
            max_v = a[i]
    return max_v

v=[17, 92, 18, 33, 58, 7, 33, 42]
print(find_max(v))
```

```python
def find_min(a):
    n=len(a)
    min_v=a[0]
    for i in range(1,n):
        if a[i] < min_v:
            min_v = a[i]
    return min_v

v=[17, 92, 18, 33, 58, 7, 33, 42]
print(find_min(v))
```

```python
def fact(n):
    f=1
    for i in range(1, n+1):
        f=f*i
        print(i,f)
    return f

print(fact(1))
print(fact(3))
print(fact(5))
print(fact(10))
# print(fact(100))
```

```python
def bubble_sort(a):
    for i in range(0,len(a)):
        for j in range(i+1,len(a)):
            if a[i]>a[j]:
                a[i],a[j]=a[j],a[i]
                print(a)
    return a

d1=[2,4,5,1,3]
d2=[7,9,5,6,3]
for d in [d1,d2]:
    print(bubble_sort(d))
```
