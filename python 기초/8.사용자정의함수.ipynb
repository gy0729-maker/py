```python
def mysum(n,a):
    result=0
    for i in range(1,n+1):
        if i % a == 0:
            result=result+i
    return result

print(mysum(100,3))
print(mysum(100,5))
```

```python
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False
    
print(is_even(10))
print(is_even(3))
```

```python
def greet(n):
    for name in n:
        msg = name + "님 환영합니다."
        print(msg)

names=['김철호','나철호','송민수']
greet(names)
```

```python
def check_number(n):
    if n>0:
        return '양수'
    elif n==0:
        return '0'
    else :
        return '음수'
    
print(check_number(5))
print(check_number(0))
print(check_number(-3))
```
