```python
class person:
    name=''
    age=0
    height=0

    def set_age(self,age):
        if age>100 or age<0:
            self.age=int(input('나이를 정확히 입력하세요:'))
        else:
            self.age=age

    def disp(self):
        print(f'이름:{self.name}, 나이:{self.age}, 키:{self.height}')

p3=person()
p3.name='김철수'
p3.set_age(-100)
p3.height=187.7
p3.disp()
```

```python
class Emp:
    name=''
    salary=0
    bonus=0
    money=0

    def __init__(self,name,salary,bonus):
        self.name=name
        self.salary=salary
        self.bonus=bonus

    def calc(self):
        self.money=self.salary*12+self.bonus

    def disp(self):
        print(f'이름:{self.name}, 월급:{self.salary}, 보너스:{self.bonus}, 연봉:{self.money}]')

e4=Emp('김철수',500,700)
e4.calc()
e4.disp()
```

```python
import math

class MyCircle:
    a=0
    b=0
    r=0
    def set_r(self,r):
        self.r=r
    def length(self):
        self.a=2*math.pi*self.r
    def area(self):
        self.b=math.pi*self.r**2
    def disp(self):
        print(f'반지름:{self.r},둘레:{self.a:.2f},넓이:{self.b:.2f}')

a=MyCircle()
a.set_r(50)
a.length()
a.area()
a.disp()
```

```python
class User:
    def __init__(self, name):
        self.name=name
    def greet(self):
        print('Hello,',self.name)

u=User('Alice')
u.greet()
```

```python
class Box:
    def __init__(self, size):
        self.size = size
    def get_size(self):
        return self.size
    
b = Box(10)
print(b.get_size())
```
