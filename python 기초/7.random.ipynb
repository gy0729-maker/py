```python
import random
#seed(기준값) 미설정시 시간기준으로 결과값이 계속 바뀜
random.seed(10)
for i in range(5):
    print(random.random())

for i in range(5):
    print(random.uniform(1,100))
```

```python
import random
food=['짜장면','짬뽕','탕수육','군만두']
f=random.choice(food)
print(f)
f=random.choices(food,k=2)
print(f)
random.shuffle(food)
print(food)
print(random.sample(food,2))
```

```python
import random
a=random.randint(1,9)
b=random.randint(1,9)
question='%d + %d = ?'%(a,b)
c=int(input(question))

if c == a+b:
    print('정답입니다.')
else:
    print('틀렸습니다.')
```

```python
import random

secret = random.randint(1,100)
count = 0
while True:
    num=int(input('숫자를 입력하세요(끝내려면 0)'))
    if num == 0:
        break
    count += 1
    if num == secret:
        print('%d번만에 맞췄습니다' % count)
        break
    elif num > secret:
        print('더 작은 수를 입력하세요.')
    else:
        print('더 큰 수를 입력하세요.')
```

```python
import random

words = {'apple':'사과','banana':'바나나','cherry':'체리'}
quiz = random.choice(list(words.keys()))
correct_meaning = words[quiz]
user_input = input(f"'{correct_meaning}'의 영어 단어는?")
if user_input == quiz:
    print('Excellent!')
else:
    print(f'아쉬워요! 정답은 {quiz}입니다.')
```
