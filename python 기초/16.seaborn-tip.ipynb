```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='time',data=tips)
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=tips)
plt.show()
```

```python
sns.barplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.barplot(y='day',x='tip',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.countplot(x='day',data=tips)
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20)
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
plt.show()
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.boxplot(x='smoker', y='tip', data=tips)
plt.show()
```

```python
sns.boxplot(x='smoker', y='tip', data=tips, hue='sex')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.violinplot(x='day',y='tip',hue='sex',data=tips,split=True)
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```

```python
day_counts=tips['day'].value_counts()
plt.pie(
    day_counts,
    labels=day_counts.index,
    autopct='%1.1f%%'
)
plt.show()
```

```python
tips=sns.load_dataset('tips')
styles=['white','dark','whitegrid','darkgrid','ticks']
for style in styles:
    sns.set_style(style)
    sns.barplot(x='day',y='tip',data=tips)
    plt.title(f'style={style}')
    plt.show()
```
