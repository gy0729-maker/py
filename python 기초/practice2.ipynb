```python
import math

class Circle:
    def __init__(self, r):
        self.volume=4/3*math.pi*(r*r*r)
        self.area=4*math.pi*(r*r)

c = Circle(5)
c.volume()
c.area()
print(f'구의반지름:',c, '구의부피:',c.volume(), '구의겉넓이:',c.area())
```
