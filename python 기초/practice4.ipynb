```python
def print_n_times(n, *values):
    for i in range(n):
        for value in values:
            print(value)
        print()

print_n_times(3, '안녕하세요', '반갑습니다.')
```

```python
def mul(*values):
    output=1
    for i in values:
        output *= i
    return output

print(mul(5,7,9,10))
```

```python
def factorial(n):
    output=1
    for i in range(1,n+1):
        output*=i
    return output

print(factorial(5))
```

```python
def flatten(data):
    output=[]
    for item in data:
        if type(item)==list:
            output+=flatten(item)
        else:
            output.append(item)
    return output

example = [[1,2,3],[4,[5,6]],7,[8,9]]
print(example)
print(flatten(example))
```

```python
dictionary = {1:1, 2:1}
def fibonacci(n):
    if n in dictionary:
        return dictionary[n]
    else:
        output = fibonacci(n-1)+fibonacci(n-2)
        dictionary[n] = output
        return output

print(fibonacci(50))
```

```python
min = 2
max = 10
people = 10
memo = {}

def result(left, min):
    key = str([left, min])
    if key in memo:
        return memo[key]
    if left<0:
        return 0
    if left==0:
        return 1
    
    count=0
    for i in range(min,max+1):
        count += result(left-i, i)
    memo[key]=count
    return count

print(result(10,2))
```

```python
min_val = 2
max_val = 10
memo = {}

def result(left, start):
    key = (left, start)
    if key in memo:
        return memo[key]
    if left == 0:
        return 1
    if start > left:          # 더 이상 만들 수 없음
        return 0

    count = 0
    upper = min(max_val, left)  # i가 left를 넘으면 의미 없음
    for i in range(start, upper + 1):
        count += result(left - i, i)

    memo[key] = count
    return count

print(result(100, 2))
print(memo)
```
