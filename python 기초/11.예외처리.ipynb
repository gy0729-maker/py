```python
print('종료하려면 q를 입력하세요.')
while True:
    num1 = input('분자: ')
    if num1 == 'q':
        break
    num2 = input('분모: ')
    if num2 == 'q':
        break

    try:
        result = (f'{int(num1)/int(num2):.2%}')
    except:
        print('분모에 0이 올 수 없습니다.')
    else:
        print(result)
print('프로그램을 종료합니다.')
```

```python
try:
    f=open('c:/data/test.txt','r')
except:
    print('파일이 없습니다.')
else:
    while True:
        line=f.readline()
        if not line:
            break
        print(line, end='')
    f.close()
```

```python
filename= 'c:/data/test.txt'  #한글인코딩방식 cp949:2byte / utf-8:3byte
with open(filename, encoding='utf-8') as file:
    contents = file.read()
    words = contents.split()
    num = len(words)
    print(f'파일이름: {filename}, 단어개수: {num}')
```

```python
filename = 'c:/data/test.txt'
try:
    with open(filename, encoding='utf-8') as file:
        contents = file.read()
except:
    msg='파일을 찾을 수 없습니다.'
    print(msg)
else:
    words=contents.split()
    num=len(words)
    print(f'파일이름: {filename}, 단어수: {num}')
```

```python
total=0
try:
    with open('c:/data/numbers.txt','r') as f:
        for line in f:
            total += float(line.strip())
        print('합계:', total)
except:
    print('잘못된 데이터')
```

```python
total=0
with open('c:/data/numbers.txt','r') as f:
    for line in f:
        try: total += int(line.strip())
        except:
            print(f'{line}: 숫자가 아닙니다.')
print('합계:', total)
```
