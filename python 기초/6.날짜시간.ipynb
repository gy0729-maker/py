```python
from datetime import date, datetime
today=date.today()
print('today: {}'.format(today))
#==print(f'today: {today}')

print('year: {0}'.format(today.year))
print('month: {0}'.format(today.month))
print('day: {0}'.format(today.day))
print('current_datetime:{0}'.format(datetime.today()))
```

```python
today=datetime.today()
print('{}'.format(today.strftime('%m/%d/%y')))
print('{}'.format(today.strftime('%y-%m-%d %H:%M:%S')))
#Y 4자리 y 2자리 H 24시간 I 12시간
print('{}'.format(today.strftime('%y-%m-%d %I:%M:%S')))

now=datetime.now()
print(now)
print('%d년 %d월 %d일'%(now.year, now.month, now.day))
print('%d:%d:%.2f'%(now.hour, now.minute, now.second))
# '%옵션' % (값)
    #d정수 f실수 s문자
```

```python
#시간을 포맷에 맞춰 출력하기
output_a = now.strftime('%Y.%m.%d %H:%M:%S')
print(output_a)
output_b = '{}년 {}월 {}일 {}시 {}분 {}초'.format(
    now.year,now.month,now.day,now.hour,now.minute, now.second)
print(output_b)
output_c = now.strftime('%y{} %m{} %d{} %I{} %M{} %S{}').format(*'년월일시분초')
print(output_c)
```

```python
import datetime
now = datetime.datetime.now()

#datetime.timedelta로 시간 더하기
after = now + datetime.timedelta(
    weeks=1, days=0, hours=1, minutes=1, seconds=1
)
print(after.strftime('%Y-%m-%d %H:%M:%S'))
```
