```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins=sns.load_dataset('penguins')
penguins
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    data=penguins
)
plt.show()
```

```python
sns.scatterplot(
    x='body_mass_g',
    y='bill_length_mm',
    data=penguins,
    hue='species'
)
plt.show()
```

```python
sns.barplot(
    x='species',
    y='body_mass_g',
    data=penguins
)
plt.show()
```

```python
sns.boxplot(
    x='sex',
    y='body_mass_g',
    data=penguins
)
plt.show()
```

```python
sns.boxplot(
    x='species',
    y='bill_length_mm',
    data=penguins
)
plt.show()
```

```python
sns.barplot(
    x='sex',
    y='flipper_length_mm',
    data=penguins
)
plt.show()
```

```python
sns.histplot(
    penguins['flipper_length_mm'],
    bins=20
)
plt.show()
```

```python
sns.regplot(
    x='flipper_length_mm',
    y='body_mass_g',
    data=penguins
)
plt.show()
```

```python
sns.pairplot(
    penguins,
    hue='species'
)
plt.show()
```
