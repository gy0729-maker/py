```python
names=['김철수','박철호','최정호','이철수','홍찬수']
points=[100,80,70,60,90]
grades=['A','B','B','C','A']
for (name,point,grade) in list(zip(names,points,grades)):
    print('이름:',name,'/ 점수:', point,'/등급:',grade)
```

```python
import random

words = {'apple':'사과','banana':'바나나','cherry':'체리'}
quiz = random.choice(list(words.keys()))
correct_meaning = words[quiz]  #딕셔너리 값추출->사전[key]


user_input = input(f"'{correct_meaning}'의 영어 단어는?")
if user_input == quiz:
    print('Excellent!')
else:
    print(f'아쉬워요! 정답은 {quiz}입니다.')
```

```python
def myabs(n):
    if n>=0:
        return n
    else:
        return -n
    
print(myabs(-10))
```

```python
def mysum(n,a):
    result=0
    for i in range(1,n+1):
        if i % a == 0:
            result=result+i
    return result
    
print(f'{mysum(10,3)}')
```
