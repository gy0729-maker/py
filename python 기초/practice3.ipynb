```python
limit=10000
i=1
sum_value=0

while sum_value<limit:
    i += 1
    sum_value+=i
print(f'{i}을 더할 때 10000을 넘으며 그때의 값은 {sum_value}입니다.')
```

```python
max_value=0
a=0
b=0

for i in range(1,51):
    j = 100-i
    if i*j>max_value:
        a=i
        b=j
        max_value=a*b
print(f'최대가 되는 경우: {a}*{b}={max_value}')
```

```python
dna = 'ctacaatgtcagtatacccattgcattagccgg'
counter={}

for i in dna:
    if i in counter:
        counter[i]+=1
    else:
        counter[i]=1

print(counter)

for key in counter:
    print(f'{key}의 개수: {counter[key]}')
```

```python
dna = 'ctacaatgtcagtatacccattgcattagccgg'
counter={}

for i in range(0,len(dna),3):
    codon =  dna[i:i+3]
    if len(codon) == 3:
        if codon in counter:
            counter[codon]+=1
        else: counter[codon]=1
print(counter)
```

```python
a=[1,2,[3,4],5,[6,7],[8,9]]
output = []

for i in a:
    if type(i) == list:
        for j in i:
            output.append(j)
    else:
        output.append(i)

print(output)
```
