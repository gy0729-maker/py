```python
import pandas as pd
df = pd.read_csv('member.csv')
df.head()
```

```python
#가로막대
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.barh(df['name'], df['age'])
plt.title('Member Age (Horizontal)')
plt.xlabel('age')
plt.ylabel('name')
plt.show()
```

```python
#세로막대-나이순
import matplotlib.pyplot as plt

sorted_df=df.sort_values('age')

plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(10,5))
plt.bar(sorted_df['name'], sorted_df['age'])
plt.xticks(rotation=45)
plt.title('Member Age (Sorted)')
plt.xlabel('name')
plt.ylabel('age')
plt.show()
```

```python
#나이별 카운트 막대그래프
age_count=df['age'].value_counts().sort_index()
#print(age_count)

plt.bar(age_count.index, age_count.values)
plt.title('Count by Age')
plt.xlabel('age')
plt.ylabel('count')
plt.show()
```

```python
#박스플롯(나이)
plt.boxplot(df['age'])
plt.title('Age Boxplot')
plt.ylabel('age')
plt.show()
```

```python
#도수분포표+수직선(평균)

plt.hist(df['age'], bins=5) #bins=구간수
plt.axvline(df['age'].mean(), linestyle='--')
plt.title('Age Distribution With Mean')
plt.xlabel('age')
plt.ylabel('count')
```
