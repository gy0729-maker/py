```python
cursor = conn.cursor()
sql = 'select * from movie_rating order by movie_id'
cursor.execute(sql)
rows = cursor.fetchall()
print('앞 5개:')
for row in rows[:5]:
    print(row)
```

```python
import pandas as pd

df = pd.DataFrame(
    rows,
    columns=[
        'movie_id',
        'duration_min',
        'genre_score',
        'story_score',
        'acting_score',
        'rating_level',
    ]
)

print('\n데이터프레임 head():')
print(df.head())
```

```python
csv_path = 'movie_rating.csv'
df.to_csv(csv_path, index=False)
print(f'\nCSV 저장 완료: {csv_path}')

cursor.close()
conn.close()
print('DB 종료')
```
