```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('c:/data/ontime/2006.csv')
df2 = pd.read_csv('c:/data/ontime/2007.csv')
df3 = pd.read_csv('c:/data/ontime/2008.csv')
df1=pd.concat([df1, df2, df3])
```

```python
df1.shape
```

```python
df1.columns
```

```python
df2 = df1[['Year','Month','ArrDelay','DepDelay']]
df2.head()
```

```python
df2.tail()
```

```python
df2.shape
```

```python
df2 = df2.dropna(how='any')
```

```python
df2.shape
```

```python
result = df2.groupby(['Year','Month'], as_index=False).count()
result.head()
```

```python
result.to_csv('c:/data/ontime/result.csv')
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    'mysql+pymysql://user:1234@localhost:3306/ontime'
)
result.to_sql(
    name='ontime_result',
    con=engine,
    if_exists='replace',
    index=False
)
```
