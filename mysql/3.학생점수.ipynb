```python
import pymysql

conn = pymysql.connect(
    host='localhost', #내컴퓨터=127.0.0.1 / ip주소가 들어갈수도있음
    user='user',
    password='1234',
    database='testdb',
    charset='utf8'
)
cursor = conn.cursor()
cursor.execute('select * from score order by student_id')
rows = cursor.fetchall()  #결과를 리스트로 저장

cursor.close()
conn.close()
```

```python
import pandas as pd
df=pd.DataFrame(
    rows,
    columns=['studend_id', 'study_hours', 'attendance', 'assignment_score', 'sleep_hours', 'final_score']
)
df.head()
```

```python
backup_path='score.csv'
df.to_csv(backup_path, index=False)
```
