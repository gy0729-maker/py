```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('movie_rating.csv')
print(df)
```

```python
sns.set_theme(style='whitegrid')

plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

plt.figure()
sns.countplot(x='rating_level', data=df)

plt.title('만족도 분포')
plt.xlabel('rating_level (0:불만,1:보통,2:만족)')
plt.ylabel('count')
plt.show()
```

```python
sns.boxplot(x='rating_level',y='story_score',data=df)
plt.title('만족도별 스토리 점수 분포')
plt.xlabel('rating_level')
plt.ylabel('story_score')
plt.show()
```

```python
sns.boxplot(x='rating_level',y='duration_min',data=df)
plt.title('만족도별 스토리 점수 분포')
plt.xlabel('rating_level')
plt.ylabel('duration_min')
plt.show()
```

```python
colors=['red','green','blue']
sns.scatterplot(
    x='acting_score',
    y='story_score',
    hue='rating_level',
    data=df,
    alpha=1.0,
    palette=colors
)
plt.title('연기점수 vs 스토리점수 (만족도 색상)')
plt.xlabel('acting_score')
plt.ylabel('story_score')
plt.show()
```
