```python
#회원조회(조건)
import pymysql

conn = pymysql.connect(
    host='localhost',
    user='user',
    password='1234',
    database='testdb',
    charset='utf8'
)

cursor = conn.cursor()
sql = 'select name, age from member where age >= %s'
cursor.execute(sql, (20,))
rows = cursor.fetchall()
for row in rows:
    print(row)
cursor.close()
conn.close()

cursor.close()
conn.close()
```

```python
#회원추가
import pymysql

conn = pymysql.connect(
    host='localhost',
    user='user',
    password='1234',
    database='testdb',
    charset='utf8'
)

cursor = conn.cursor()
sql = 'INSERT INTO member (name, age) VALUES (%s, %s)'

cursor.execute(sql, ('최희선', 25))
conn.commit()
print('저장되었습니다.')

sql = 'select * from member where name = %s'
cursor.execute(sql, ('최희선',))
print(cursor.fetchone())

cursor.close()
conn.close()
```

```python
#회원수정
#cmd -> update member set age=21 where name='김민수';

cursor = conn.cursor()
sql = 'update member set age=%s where name=%s'
cursor.execute(sql, (21, '김민수'))
conn.commit()

sql = 'select * from member where name = %s'
cursor.execute(sql, ('김민수',))
print(cursor.fetchone())

cursor.close()
conn.close()
```

```python
#회원삭제
#delete from member where name='김민수;

cursor = conn.cursor()
sql = 'delete from member where name = %s'
cursor.execute(sql, ('김민수',))
conn.commit()

sql='select * from member where name = %s'
cursor.execute(sql, ('김민수',))
print(cursor.fetchone())

cursor.close()
conn.close()
```

```python
#회원 테이블을 csv파일로 저장

import pymysql

conn = pymysql.connect(
    host='localhost',
    user='user',
    password='1234',
    database='testdb',
    charset='utf8'
)

cursor = conn.cursor()
cursor. execute('select * from member order by id') #id기준 오름차순정렬 / 내립차순 +desc
rows = cursor.fetchall()

cursor.close()
conn.close()
```

```python
import pandas as pd
df = pd.DataFrame(rows, columns=['id','name','age'])
df
```

```python
backup_path = 'member.csv'
df.to_csv(backup_path, index=False)
```
